## 3. Создать файл отчета по выполнению практической работы и поместить его в репозиторий




1. Для выполнения данного задания необходимо создать и скачать файл отчета Google colab

image.png

2. Далее необходимо поместить его в созданную ранее папку и с помощью команды git add (название файла с расширением)  добавить файл в git, и команды git commit -m "Сообщение" сделать снимок этого изменения.В итоге мы получаем такой результат.

image.png
image.png

## 4. Зафиксируйте в отчете все проделанные шаги и создайте версию отчета в GIT.

1. Для выполнения данного задания необходимо зафиксировать все ранее созданные действия в отчете, скачать его и копировать его с заменой в папку git-a. Далее с помощью команды git status проверяем наш файл(должен быть статус modifier), добавляем его с помощью git add .  , и с помощью git commit -m 'сообщение' делаем его снимок

image.png

## 5. Создайте ветку отчета в GIT, добавьте описания шагов в отчет и создайте версию отчета в новой ветке.

1. Для выполнения данного задания необходимо воспользоваться командой git branch <имя ветки>, в результате чего у нас создастся новая ветка. Чтобы вывести все существующие ветки, необходимо воспользоваться командой git branch -a

image.png

2. С помощью команды git checkout <название ветки> переключиться на нее. Далее создаем новую папку, куда в последствии добавим новую версию отчета. С помощью команды 'git add .' добавляем папку с нашим отчетом в новую ветку  git-a, и с помощью git commit -m "Сообщение", делаем снимок наших изменений.

image.png
